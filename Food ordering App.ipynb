```python
class Restaurant:
    def __init__(self):
        self.food = {}
        self.food_id = len(self.food)+1
        self.ordered_item = []
        self.personal_details = {}
#Admin page    
    def add_food_item(self):
        self.name = input("Enter the name of the food: ")
        self.quantity = input("Enter the quantity: ")
        self.price = int(input("Enter the price: "))
        self.discount = input("Enter the discount: ")
        self.stock = input("Enter the available stock of the item: ")
        self.new_item = {"Name":self.name, "quantity":self.quantity, "price":self.price, "discount":self.discount, "stock":self.stock}
        self.food_id = len(self.food) + 1
        self.food[self.food_id] = self.new_item
        
    def edit_food_item(self):
        f_id = int(input("Enter the food id ypu want to update: "))
        for i in self.food[f_id]:
            self.food[f_id] = input("Enter the" + i + " you want to update: ")
        print("Food item is updated successfully...!")
    
    def all_food_items(self):
        print("List of all food items are: ")
        for i in self.food:
            print("Food id: ",i,"\n")
            for j in self.food[i]:
                print(j, ":", self.food[i][j])
        
    def remove_item(self):
        del self.food[int(input("Enter the food id you want to remive from the menu: "))]
        print("The food item is deleted successfully")
        
        
#User page:
    
    def register(self):
        self.fullname = input("Full name: ")
        self.phone = int(input("Phone number: "))
        self.email = input("Email id: ")
        self.address = input("Address: ")
        self.password = input("Password: ")
        self.personal_details = {"fullname":self.fullname, "phone":self.phone, "email": self.email, "address":self.address, "password":self.password}
        print("You are registered successfully...!")

    def login(self):
        while True:
            user_id = input("Enter your user id: ")
            password = input("Enter your password: ")
            if user_id == self.email:
                if password == self.password:
                    print("Logged in successfully...!")
                    break
                else:
                    print("Invalid password")
            else:
                print("Invalid user id")
                
    def home(self):
        select = input("Login | Signup: ")
        if select.lower() == "login":
            self.login()
            print("Welcome %s." %self.fullname)
            print("\t 1. Place Order\n"
                  "\t 2. Order History\n"
                  "\t 3. Update profile")
            user_input = input("Select any of these(1,2 or 3): ")
            if user_input == "1":
                self.place_order()
            elif user_input == "2":
                self.order_history()
            elif user_input == "3":
                self.edit_profile()
        elif select.lower() == "signup":
            self.register()
            self.login()
            print("Welcome %s." %self.fullname)
            print("\t 1. Place Order\n"
                  "\t 2. Order History\n"
                  "\t 3. Update profile")
            user_input = input("Select any of these(1,2 or 3): ")
            if user_input == "1":
                self.place_order()
            elif user_input == "2":
                self.order_history()
            elif user_input == "3":
                self.edit_profile()
        else:
            print("Select any one of the option")
            

    def place_order(self):
        list_of_food = {1:"Tandoori Chicken - 4 pieces - INR 240",2:"Vegan Burger - 1 Piece - INR 320", 3:"Truffle Cake - 500 gm - INR 900"}
        print("Menu List")
        print("\t (1) Tandoori Chicken - 4 pieces - INR 240 \n"
              "\t (2) Vegan Burger - 1 piece - INR 320\n"
              "\t (3) Truffle Cake - 500gm - INR 900\n")
        user_input = int(input("what do you love to have: "))
        if user_input == 1:
            print("Your order fo Tandoori Chicken is confirmed...!")
            self.ordered_item.append(list_of_food[user_input])
        elif user_input == 2:
            print("Your order fo Vegan Burger is confirmed...!")
            self.ordered_item.append(list_of_food[user_input])
        elif user_input == 3:
            print("Your order fo Truffle Cake is confirmed...!")
            self.ordered_item.append(list_of_food[user_input])
            
    def order_history(self):
        for i in self.ordered_item:
            print(i)
            
    def edit_profile(self):
        key = str(input("What you want to update: "))
        self.personal_details[key] = input("Enter the new value: ")
        print("Profile updated successfully...!",self.personal_details)
```
